**while loop**

- while loop:- if youwant to run infinite iterations will choose while loop
- every loop we needs three things
  - initial point(starting point)
  - increment or decrement
  - condition(stop point)

- in for loop all three things mentioned in a single line: range(start,stop,step)
- in while loop we will use thses three things, in as a three lines


```python
<start point>
while <how to stop>:
    codeline1
    codeline2
    <increment or decrement>
```

```python
import random
sum_e,count_e,sum_o,count_o=0,0,0,0
i=1
while(i<5):
    num=random.randint(1,10)
    if num%2==0:
        sum_e=sum_e+num
        count_e+=1
    else:
        sum_o+=num
        count_o+=1
    i=i+1
print("even sum: ", sum_e)
print("odd sum: ",sum_o)
print("even count: ",count_e)
print("odd count: ",count_o)
```
