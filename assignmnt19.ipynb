```python
curpath=os.getcwd()
folder_name='manya'
newdir=os.path.join(curpath,folder_name)
os.makedirs(newdir)
```

```python
#3) Write the file
#     check the mode ='a'  
#    check the mode='w'
#    check the mode='x'
```

```python
fname='luckyros.txt'
path=os.path.join(newdir,fname)
with open(path, mode='w') as file:
    print(file.write('hi mummy.............'))
```

```python
with open(path, mode='r') as file:
    print(file.read())
```
