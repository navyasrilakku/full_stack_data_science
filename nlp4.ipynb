**some dis advantages of bag of words**
- it won't provide any semantic information about the words. it only gives how many times has word occurred in a  sentence and not its location or correlation with other words in the sentence
- it gives equal importance to all the words in the sentence. hence it is most useful for simple processess
- there are other methods like Tfidf, word2Vec which are more complex and useful than BOW's

  - BOW
  - tf-idf
  - word2Vec
  - Glove

**Term Frequency**
- the term is a frequency measur of a word w in a document (text) d.
- it is equal to the number of instances of word w in documnt d divided by the total number of words in document d
- term frequency serves as a metric to determine a word occurence in a document as compared to the total number of words in a document. the denominator is always same
   term frquency= no of instances of a word w in document d/ total no of words in a document d
**Inverse document frequency**
IDF= log(total no of documents in text corpus d/no of documents containing w )

```python
doc1="going rain today."
doc2="today going outside."
doc3="going watch season premiere."
doc=[doc1,doc2,doc3]
print(doc)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vector=TfidfVectorizer()
X=vector.fit_transform(doc)
X
```

```python
vector.vocabulary_
```
