```python
df=pd.read_csv(r"D:\full stack data science and ai\data sets\android-games.csv")
```

```python
df['category'].value_counts()
```

```python
text=[" ".join (cat.split()[1]) for cat in df.category]
from wordcloud import WordCloud
wc=WordCloud(height=2000, width=2000, background_color='black')
wc=wc.generate(text)
plt.imshow(wc)
plt.title('most common words in positive customer comments')
plt.axis('off')
plt.show()
```
