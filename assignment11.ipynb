```python
# Q15) string1='hello hello hello how how how how are you'
# What is the most repeated word : how
string1='hello hello hello how how how how are you'
k=string1.split(' ')
count=0
l=[]
for i in k:
    if i not in l:
        l.append( count(i))
print(l)
```
