**sets**

- sets are an unordered collection of unique elements

**properties of sets**

- unordered
         - set doesn't maintain element order 
- unique
         - set doen't allow duplicates
- mutable
         - set can change their elements

**Representation**

```python
myset=set([1,2,3,4,2,3])
myset
```

```python
set1={1,3,5,2,7,5,4,9,0,4,6,7}
set1
```

```python
list1=[1,5,3,7,8,9,2,0,3]
set1=set(list1)
set1
```

**operations on sets**

- union
- intersection
- difference
- symmetric difference

**union**
- union is used to combine two sets with out duplicate elements

```python
a={1,2,3}
b={3,4,5}
print(a|b)
```

**intersection**
- intersection is used to combine two sets and print the output only common elements in two sets

```python
a={1,2,3}
b={3,4,5}
print(a&b)
```

**difference**
- difference is used to print the values between two sets by subtracting .
- if a set contains three values 1,2,3
- another set contains three values 3,4,5
- while making difference between two sets we get the values as 1,2

```python
a={1,2,3}
b={3,4,5}
print(a-b)
```

- in case if we do b-a we got the values 4 and 5

```python
a={1,2,3}
b={3,4,5}
print(b-a)
```

**symmetric difference**
- symmetric difference is used to remove the common elements and print remaining elements

```python
a={1,2,3}
b={3,4,5}
print(a^b)
```
