```python
# Q5 from above output
# you need to extract all the emails
path=r'mbox-short.txt'
file=open(path)
list1=[]
a=[]
for i,j in enumerate(file):
    if j.startswith('From'):
        list1.append(j)
for i,j in enumerate(list1):
    if i%2!=0:
        a.append(j.rstrip('\n').lstrip('From').lstrip(':').strip())
print(a,end=' ')
```

```python
# Q6 unique emails
u=[]
for i,j in enumerate(a):
    if j not in u:
        if j in a:
            u.append(j)
print(u,end=' ')
```

```python
# Q7 you need to read the text file
path=r'emails.txt'
file=open(path)
fname=[]
lname=[]
email=[]
company=[]
for i,j in enumerate(file):
    a=j.split(';')
for i in a:
    name,other=(i.split(','))
    fname.append(name)
    a,b=(other.split('<'))
    lname.append(a)
    email2=(b.strip(">"))
    email.append(email2)
    df=email2.find('@')
    df2=email2.find('.',df)
    company.append(email2[df+1:df2])
print(fname,end=' ')
print('-------------------------------------')
print(lname,end=' ')
print('-------------------------------------')
print(email,end=' ')
print('-------------------------------------')
print(company,end=' ')
```
