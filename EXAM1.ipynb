```python
# 1. Write a program for all arithmetic operations using eval function.
num1=eval(input("enter a value of number1: "))
num2=eval(input("enter a value of number2: "))
print(f"addition of {num1} and {num2} is {num1+num2}")
print(f"subtraction of {num1} and {num2} is {num1-num2}")
print(f"multiplication of {num1} and {num2} is {num1*num2}")
print(f"division of {num1} and {num2} is {num1/num2}")
```

```python
#  3 Explain slicing and indexing with examples.
```


** slicing **
- slicing method is used to make different operations like to print numbers with a specific condition.
- it will be used in lists
- the format of slicing is---------- variable[1:100:2]
- here the first step shows that the starting value 
- the second step represents the ending value
- the third step represents the step value  that it shows the numbers based on that sttep value


```python
#example
List = [10, 20, 30, 40, 50, 60, 70]
print(List[0:8:3])
```

** indexing**
- [10,20,30,40,50]
-   1  2  3  4  5
- indexing is just lik a storage space based on that index values we can extract the required data.


```python
List=[10,20,30,40]
k=List[1]
print(k)
```

```python
## 5  Explain type casting with fundamental datatypes with example.
```


**** type casting**
- it is used to convert the data from one data type to another data type

- we can convert Int(float,str,bool)
- float(int, str,bool)
- str(int, float,bool)

in integer************

```python
print(float(10))
print(str(10))
print(bool(10))
print(type(10))
```

in float**********

```python
print(int(10.0))
print(str(10.0))
print(bool(10.0))
print(type(10.0))
```

in string**

```python
print(int('10'))
print(float('10'))
print(bool('10'))
print(type('10'))
```

in bool**

```python
print(float(True))
print(str(True))
print(int(True))
print(type(True))
```
