```python
import random
sum_e,count_e,sum_o,count_o=0,0,0,0
i=1
while(i<5):
    num=random.randint(1,10)
    if num%2==0:
        count_e+=1
        print(f"{num}")
    else:
        count_o+=1
        print(f"{num}")
    i=i+1

print("even count: ",count_e)
print("odd count: ",count_o)
```

```python
import random
sum_e,count_e,sum_o,count_o=0,0,0,0
i=1
while(i<5):
    num=random.randint(1,10)
    if num%2==0:
        sum_e=sum_e+num
        count_e+=1
        print(f"{num}")
    else:
        sum_o+=num
        count_o+=1
        print(f"{num}")
    i=i+1
print("even sum: ", sum_e)
print("odd sum: ",sum_o)
print("even count: ",count_e)
print("odd count: ",count_o)
```
