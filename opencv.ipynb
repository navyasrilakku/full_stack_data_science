**convert color image Gray**
- color image has RGB channel
- Gray means black and white
- cvtColor
- OpenCV by default take as BGR channel
- inside we ned to pass a method BGR2GRAY

```python
import numpy as np
import cv2
image= cv2.imread("photomini.jpg")
gray_img= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
gray_img.shape, image.shape
```

```python
import matplotlib.pyplot as plt
plt.subplot(1,2,1).imshow(image)
plt.subplot(1,2,2).imshow(gray_img)
```
